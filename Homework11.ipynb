```python
import math
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
fx = -5*x**2 + 10*x - 150
fx
```

```python
p = plot(fx, (x,-10,10))
```

```python
sqrt_1,sqrt_2 = solve(fx)
print(sqrt_1, sqrt_2)
```

```python
pr = diff(fx)
print(pr)
```

```python
apex = solve(diff(fx))[0]
print(apex)
```

```python
solve(diff(fx) > 0)
```

```python
solve(diff(fx) < 0)
```

```python
solve(fx < 0)
print(solve (fx < 0))
```

```python
solve(fx > 0)
print(solve (fx > 0))
```
